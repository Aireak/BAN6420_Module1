```python
import random
import string


# function to generate random factory worker names

def develope_names():
    return "".join(random.choices(string.ascii_uppercase, k=7))

# create a list of atleast 421 factory workers
factory_workers = []
for i in range(421):
    factory_worker = {
        'staff_id':  [i + 1],
        'staff_name': develope_names(),
        'salary': random.randint(5500, 35500),
        'gender': random.choice(['Male', 'Female'])
    }
    factory_workers.append(factory_worker)


# Generate worker payment slips
for factory_worker in factory_workers:
    try:
        Staff_id = factory_worker['staff_id']
        Staff_Name = factory_worker['staff_name']
        Salary = factory_worker['salary']
        Gender = factory_worker['gender']
        Employee_Level = 'L1' # tHIS IS A DEFAUCLT LEVEL

        if 10000 < Salary < 20000:
            Employee_Level = 'A1'
        if 7500 < Salary < 30000 and Gender == 'Female':
            Employee_Level = 'A5-F'
        
        print(f"Payment Slip for Factory Staff ID:{Staff_id}, Staff Name: {Staff_Name}, Staff Salary: ${Salary}, Gender: {Gender}, Staff Employee Level: {Employee_Level}")

        
    except Exception as e:
        print(f"An error occurred while processing  Staff ID: {Staff_id}. Error: {str(e)}")
```
